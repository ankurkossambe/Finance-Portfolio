```python
# Install yfinance (only needed in Colab or first run)
!pip install yfinance

# Import libraries
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# Define two NSE stock tickers
ticker1 = "TCS.NS"
ticker2 = "INFY.NS"

# Fetch last 6 months of data
data1 = yf.Ticker(ticker1).history(period="6mo")
data2 = yf.Ticker(ticker2).history(period="6mo")

# Plot both closing prices
plt.figure(figsize=(12, 6))
plt.plot(data1['Close'], label=ticker1)
plt.plot(data2['Close'], label=ticker2)
plt.title("Stock Price Comparison - Last 6 Months")
plt.xlabel("Date")
plt.ylabel("Closing Price (INR)")
plt.legend()
plt.grid()
plt.show()

# OPTIONAL: Calculate daily returns (advanced feature)
data1['Daily Return'] = data1['Close'].pct_change()
data2['Daily Return'] = data2['Close'].pct_change()

#  Export both datasets to CSV for Excel or Google Sheets
data1.to_csv("TCS_data_with_returns.csv")
data2.to_csv("INFY_data_with_returns.csv")

print(" CSV files exported! You can now download them and open in Excel.")
```
