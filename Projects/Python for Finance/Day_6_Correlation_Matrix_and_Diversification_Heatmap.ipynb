```python
# Day 6: Correlation Matrix and Diversification Heatmap

# Install required libraries (only once in Colab)
!pip install yfinance seaborn

# Import necessary libraries
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Define your portfolio
portfolio = ["TCS.NS", "INFY.NS", "HDFCBANK.NS", "RELIANCE.NS", "ITC.NS"]

# Create an empty DataFrame for prices
price_data = pd.DataFrame()

# Download historical close prices for each stock
for ticker in portfolio:
    data = yf.download(ticker, period="6mo")["Close"]
    price_data[ticker] = data

# Drop rows with missing values (NaNs)
price_data.dropna(inplace=True)

# Calculate daily returns
daily_returns = price_data.pct_change().dropna()

# Compute correlation matrix
correlation_matrix = daily_returns.corr()

# Print correlation matrix
print("Correlation Matrix:")
print(correlation_matrix.round(2))

# Plot the heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", linewidths=0.5, fmt=".2f")
plt.title("Diversification Heatmap  Stock Correlation")
plt.show()

# Export correlation matrix to CSV
correlation_matrix.to_csv("Day6_Correlation_Matrix.csv")
print("Correlation matrix saved as 'Day6_Correlation_Matrix.csv'")
```
