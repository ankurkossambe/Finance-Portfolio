```python
# Day 7: Portfolio Allocation Pie Chart with Sector Labels

# Import libraries
import matplotlib.pyplot as plt
import pandas as pd

# Define your portfolio with investment amounts and sectors
portfolio = pd.DataFrame({
    "Stock": ["TCS", "INFY", "HDFCBANK", "RELIANCE", "ITC"],
    "Investment": [50000, 30000, 20000, 40000, 15000],
    "Sector": ["IT", "IT", "Banking", "Energy", "FMCG"]
})

# Calculate percentage of each investment
portfolio["Percent"] = round((portfolio["Investment"] / portfolio["Investment"].sum()) * 100, 2)

# Set colors based on sector
sector_colors = {
    "IT": "#1f77b4",
    "Banking": "#2ca02c",
    "Energy": "#ff7f0e",
    "FMCG": "#9467bd"
}

# Map colors to stocks
portfolio["Color"] = portfolio["Sector"].map(sector_colors)

# Plot pie chart
plt.figure(figsize=(8, 8))
plt.pie(
    portfolio["Investment"],
    labels=[f"{stock}\n{amount}\n{pct}%" for stock, amount, pct in zip(portfolio["Stock"], portfolio["Investment"], portfolio["Percent"])],
    colors=portfolio["Color"],
    startangle=140,
    wedgeprops={"edgecolor": "black"},
    textprops={"fontsize": 10}
)

plt.title("Portfolio Allocation by Stock and Sector", fontsize=14)
plt.tight_layout()
plt.show()

# Export to CSV
portfolio.to_csv("Day7_Portfolio_Pie_Sectors.csv", index=False)
print("Pie chart displayed and data exported to 'Day7_Portfolio_Pie_Sectors.csv'")
```
